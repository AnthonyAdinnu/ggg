```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Summing the 'Amount' for each 'Category' to prepare data for the bar chart
category_sales = df.groupby('Category')['Amount'].sum()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
category_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales Amount per Category')
plt.xlabel('Category')
plt.ylabel('Total Sales Amount (INR)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
y = np.array([500, 1000, 29000, 2000, 7000, 18500])
mylabels = ["Bottom", "Ethnic Dress", "Set", "Top", "Western Dress", "Kurta"]
plt.title('Segments showing Total Sales Amount per Category')
plt.pie(y, labels = mylabels)
plt.show() 
```

```python
x = np.array([500, 1000, 29000, 2000, 7000, 18500])
y = np.array(["Bottom", "Ethnic Dress", "Set", "Top", "Western Dress", "Kurta"])
plt.title('Data Points for Total Sales Amount on X axis')
plt.scatter(x, y)
plt.show()
```
